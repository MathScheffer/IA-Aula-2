```python
import heapq
def busca_star(origem, destino, conexoes, heuristica):
    """
    Encontra o caminho mínimo entre duas cidades usando o algoritmo A*.

    Args:
        origem: A cidade de origem.
        destino: A cidade de destino.
        conexoes: Dicionário de conexões entre as cidades.
        heuristica: Dicionário de estimativas de distância para cada cidade.

    Returns:
        Uma lista de cidades representando o caminho percorrido.
    """
    fronteira = [(origem, 0)] # [(cidade, custo total)]
    visitados = set() # Conjunto de cidades visitadas
    custo_do_caminho = {origem: 0} # Custo para chegar em cada cidade
    caminho = {} # Dicionário para armazenar o caminho percorrido

    while fronteira:
        atual, _ = heapq.heappop(fronteira) # Pega o nó com menor custo total
        visitados.add(atual)

        if atual == destino:
            return reconstruir_caminho(atual, caminho)

        for vizinho, distancia in conexoes.get(atual, []):
            if vizinho not in visitados:
                custo = custo_do_caminho[atual] + distancia + heuristica[vizinho]
                custo_do_caminho[vizinho] = custo
                heapq.heappush(fronteira, (vizinho, custo))

# Reconstrui o caminho a partir do destino
def reconstruir_caminho(destino, caminho):
    caminho_revertido = []
    while destino is not None:
        caminho_revertido.append(destino)
        destino = caminho[destino]
    caminho_revertido.reverse()
    return caminho_revertido

# Executa a busca do caminho
busca_star('A', 'B', conexoes, heuristica)
```
